```python
# Importamos librerias
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Definimos driver
driver = wb.Chrome()
```

```python
# Funcion para scrapear links de la pagina
def get_car_links(driver):
    car_elements = driver.find_elements(By.XPATH, "//h2[@class='poly-box poly-component__title']//a")
    car_links = [element.get_attribute("href") for element in car_elements]
    return car_links
```

```python
# Funcion para pasar de pagina
def go_to_next_page(driver):
    try:
        #next_button = driver.find_element(By.XPATH, "//li[@class = 'andes-pagination__button andes-pagination__button--next']/a")
        next_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//li[@class = 'andes-pagination__button andes-pagination__button--next']/a")))
        ActionChains(driver).move_to_element(next_button).click().perform()
        return True
    except:
        return False
```

```python
# Definir funcion principal
def scrape_all_car_links(start_url, max_paginas):
    driver.get(start_url)
    all_car_links = []
    pagina = 0

    while True:
        # Scrapear links de la pagina actual
        car_links = get_car_links(driver)
        all_car_links.extend(car_links)
        
        # Contar paginas
        pagina += 1

        # Parar tras un cierto numero de paginas
        if pagina >= max_paginas:
            print(f"Se llego al maximo de paginas: {max_paginas}")

        # Intentar pasar a la proxima pagina, si es posible
        if not go_to_next_page(driver):
            print(f"No hay mas paginas, se llego hasta: {pagina}")
            break
        
        # Esperamos unos segundos
        time.sleep(3)
    
    return all_car_links
```

```python
# Definir URL por el que comenzar y numero maximo de paginas
start_url = "https://autos.mercadolibre.com.ar/autos-usados"
max_paginas = 100
```

```python
# Obtener links de autos
all_car_links = scrape_all_car_links(start_url, max_paginas)
```

```python
# Printear cantidad de links scrapeados
print(len(all_car_links))
```

```python
# Printear todos los links scrapeados
for link in all_car_links:
    print(link)
```

```python
# Cerrar sesion de navegador
driver.quit()
```

```python
# Guardar lista de links en un pickle
import pickle
with open('car_links.pkl', 'wb') as f:
    pickle.dump(all_car_links, f)
```
